```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python

# Specify the correct path and filename
filename = '../../data/obesity_final.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(filename, sep=',')

# Display the DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Podemos ver que el promedio de edad de los involucrados en el estudio de obesidad es de 20 años de edad. 

```python
# Bar plot of Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Según el gráfico anterior, podemos ver que la mayoría de las personas que asistieron al estudio tienen una distribución de proporciones iguales entre hombres y mujeres. Lo cual nos dice que no hay ningún cesgo entre género.

```python
# Box plot of Weight by NObeyesdad
plt.figure(figsize=(12, 8))
sns.boxplot(x='nobeyesdad', y='weight', data=df)
plt.title('Weight Distribution by Obesity Level')
plt.xlabel('Obesity Level')
plt.ylabel('Weight')
plt.xticks(rotation=45)
plt.show()
```

Este gráfico es un diagrama de caja y bigotes (también conocido como boxplot), que muestra la distribución del peso de los individuos clasificados por niveles de obesidad. En un diagrama de caja y bigotes, la "caja" representa el rango intercuartílico (IQR, por sus siglas en inglés), que abarca del primer cuartil (Q1) al tercer cuartil (Q3). La "línea" dentro de la caja marca la mediana (Q2). Los "bigotes" se extienden desde la caja hasta el valor más alto y el más bajo dentro de 1.5 veces el IQR, y los puntos fuera de este rango se consideran valores atípicos, lo que significa que son inusualmente altos o bajos.

En este gráfico específico, hay siete categorías de peso:

- Normal Weight (Peso Normal)
- Overweight Level I (Sobrepeso Nivel I)
- Overweight Level II (Sobrepeso Nivel II)
- Obesity Type I (Obesidad Tipo I)
- Insufficient Weight (Peso Insuficiente)
- Obesity Type II (Obesidad Tipo II)
- Obesity Type III (Obesidad Tipo III)
- Cada categoría tiene su propio boxplot, y podemos observar cómo se distribuyen los pesos para cada nivel de obesidad. Por ejemplo, el rango de pesos en "Normal - Weight" es más bajo que en "Obesity Type I". Los puntos que aparecen como círculos fuera de los bigotes representan valores atípicos dentro de esa categoría específica.









```python
# Scatter plot of Height vs. Weight
plt.figure(figsize=(10, 6))
sns.scatterplot(x='height', y='weight', data=df, hue='gender')
plt.title('Height vs. Weight by Gender')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

Este gráfico es un diagrama de dispersión que representa la relación entre la altura y el peso, diferenciado por género. Los puntos azules representan a las mujeres (Female) y los puntos naranjas representan a los hombres (Male). La altura está en el eje horizontal (X) y el peso en el eje vertical (Y).

El gráfico muestra cómo el peso varía con la altura para cada género y puede ayudar a visualizar si hay una tendencia o patrón en la relación entre estas dos variables. Por ejemplo, podría ser que a mayor altura, mayor peso, lo que sería una correlación positiva. Además, podemos observar si hay diferencias notables en la distribución del peso y la altura entre géneros. Por ejemplo, si los puntos naranjas tienden a estar más altos en el eje Y a la misma altura, eso indicaría que los hombres de ese rango de altura tienden a pesar más que las mujeres.

Además, cualquier agrupamiento de puntos o la presencia de valores atípicos (puntos que están lejos del resto de los datos) pueden ser de interés para quienes estudian la relación entre estas dos variables.








```python
# Correlation matrix heatmap
numerical_data = df[['age', 'height', 'weight']] # Add more numerical columns if necessary
correlation_matrix = numerical_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

Este gráfico representa una matriz de correlación, la cual muestra el coeficiente de correlación entre pares de variables numéricas. El coeficiente de correlación mide la relación lineal entre dos variables y puede variar de -1 a +1. Un valor de 1 significa una correlación positiva perfecta, -1 significa una correlación negativa perfecta y 0 indica que no hay correlación.

En la matriz, cada celda muestra el coeficiente de correlación entre las variables de la fila y la columna:

- "age" (edad) y "height" (altura) tienen una correlación de -0.035, lo que sugiere una relación negativa muy débil, casi inexistente, entre la edad y la altura.
- "age" y "weight" (peso) tienen una correlación de 0.2, lo que indica una relación positiva débil, sugiriendo que a medida que la edad aumenta, podría haber una tendencia leve hacia el aumento de peso.
- "height" y "weight" tienen una correlación de 0.45, lo que sugiere una correlación moderadamente positiva. Esto significa que a mayor altura, generalmente se observa un mayor peso.
- Las celdas diagonales siempre muestran una correlación de 1, porque son la correlación de cada variable consigo misma. Los colores ayudan a visualizar la fuerza y la dirección de la correlación: el rojo representa una correlación positiva, el azul una correlación negativa y el blanco (o gris claro) ninguna o una correlación muy débil.








```python

# Assuming 'df' is your DataFrame and it includes a column 'NCP' which represents the number of main meals.
ncp_counts = df['ncp'].value_counts()

# Simple Histogram
plt.hist(df['ncp'], bins=range(int(df['ncp'].min()), int(df['ncp'].max()) + 1, 1), edgecolor='black')
plt.title('Distribution of Number of Main Meals (NCP)')
plt.xlabel('Number of Main Meals')
plt.ylabel('Frequency')
plt.show()
```

Este gráfico es un histograma que muestra la distribución de la frecuencia del número de comidas principales (NCP) que las personas tienen en un día. Cada barra representa el número de veces (frecuencia) que un determinado número de comidas principales ha sido reportado en un conjunto de datos.

La escala horizontal (eje X) muestra el número de comidas principales, dividido en intervalos (1.0, 1.5, 2.0, etc.), y la escala vertical (eje Y) muestra la frecuencia con la que cada intervalo ocurre en el conjunto de datos.

Por ejemplo, la barra más alta representa que la mayoría de las personas reportaron tener 3 comidas principales al día. Las barras más bajas para 1 y 2 comidas principales indican que hay menos personas que reportan tener solamente 1 o 2 comidas principales al día.

Este tipo de gráfico es útil para entender rápidamente cómo se distribuye una variable en un conjunto de datos, en este caso, el número de comidas principales que las personas suelen tener.








```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume df is your DataFrame and 'NCP' is the column with the number of main meals.

# Define the bins for the categories
bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]  # Adjust these values based on your specific needs.
# Create labels for the bins
labels = ['1-2', '2-3', '3-4', '4-5', '5+']

# Categorize the data into bins
df['NCP_category'] = pd.cut(df['ncp'], bins=bins, labels=labels, right=False)

# Count the occurrences of each category
ncp_counts = df['NCP_category'].value_counts(normalize=True) * 100  # Get percentage counts

# Sort the counts by index to maintain the order of the bins
ncp_counts = ncp_counts.sort_index()

# Plot the pie chart
fig, ax = plt.subplots()  # Create a figure and an axes.
wedges, texts, autotexts = ax.pie(ncp_counts, labels=ncp_counts.index, autopct='%1.1f%%', startangle=140)

# Title for the pie chart
plt.title('Percentage of Number of Main Meals (NCP)')

# Add a legend
ax.legend(wedges, labels,
          title="Meal Ranges",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

Este gráfico es un gráfico de pastel o de tarta que muestra la distribución porcentual del número de comidas principales (NCP) que las personas informaron tener. Cada sección del pastel representa un rango de número de comidas principales y está codificado con colores que corresponden a la leyenda en el lado derecho del gráfico:

- El rango de 1-2 comidas está representado en azul con un 14.9% del total.
- El rango de 2-3 comidas está en naranja con un 8.2%.
- El rango de 3-4 comidas es el más grande, en verde, con un 69.4%.
- El rango de 4-5 comidas está en rojo con un 7.5%.
- El rango de 5 o más comidas está en púrpura con un 0.0%, indicando que probablemente no hubo reportes de personas que comieran 5 o más comidas principales en un día.
- El gráfico de pastel es útil para entender la proporción relativa de diferentes categorías dentro de un conjunto de datos. En este caso, podemos ver rápidamente que la mayoría de las personas informaron tener entre 3 y 4 comidas principales al día, ya que esta categoría ocupa la mayor parte del gráfico.
